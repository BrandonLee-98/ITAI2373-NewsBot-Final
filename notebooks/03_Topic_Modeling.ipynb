## Objective
This notebook focuses on **Module A: Topic Discovery**. 
We will implement a machine learning pipeline to:
1. Convert news text into a mathematical Document-Term Matrix.
2. Apply **Latent Dirichlet Allocation (LDA)** to discover latent themes.
3. Visualize the top keywords per topic to provide 'Behind the Scenes' insights.

```python
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import pandas as pd

# Sample corpus representing different news beats
corpus = [
    "The court ruled that the amendment protects individual privacy in digital spaces.",
    "Legal experts say the liability framework for AI developers is still evolving.",
    "Artificial intelligence models are becoming more efficient at reasoning and logic.",
    "OpenAI and Google are competing for dominance in the large language model market.",
    "The Federal Reserve is monitoring inflation and interest rate trends closely.",
    "Stock markets rallied as investors gained confidence in the economic soft landing."
]

print("✅ Corpus ready for statistical modeling.")
```

```python
# 1. Vectorization: Convert text to token counts (excluding English stop words)
vectorizer = CountVectorizer(stop_words='english')
dtm = vectorizer.fit_transform(corpus)

# 2. Initialize and Fit LDA
# We seek 3 topics: likely Legal, Tech, and Finance based on our corpus
lda = LatentDirichletAllocation(n_components=3, random_state=42)
lda.fit(dtm)

print("✅ LDA Model trained on news corpus.")
```

```python
# 3. Visualization of Hidden Topics
words = vectorizer.get_feature_names_out()

def plot_top_words(model, feature_names, n_top_words, title):
    fig, axes = plt.subplots(1, 3, figsize=(15, 6), sharex=True)
    axes = axes.flatten()
    for topic_idx, topic in enumerate(model.components_):
        top_features_ind = topic.argsort()[: -n_top_words - 1 : -1]
        top_features = [feature_names[i] for i in top_features_ind]
        weights = topic[top_features_ind]

        ax = axes[topic_idx]
        ax.barh(top_features, weights, color='navy')
        ax.set_title(f"Topic {topic_idx + 1}", fontsize=14)
        ax.invert_yaxis()
        ax.tick_params(axis='both', which='major', labelsize=12)
    
    plt.suptitle(title, fontsize=18)
    plt.subplots_adjust(top=0.85, bottom=0.05, wspace=0.90)
    plt.show()

plot_top_words(lda, words, 5, "Top Keywords per Discovered News Topic")
```

## Findings
The visualization above clearly shows the 'clusters' of keywords that define each topic. 
By identifying these clusters, NewsBot can automatically group similar articles or suggest related content to the user, satisfying the requirement for **Content Clustering using topic distributions**.
