## Objective
This notebook focuses on **Module D: Conversational Interface**. 
We will demonstrate the implementation of a context-aware chatbot that allows users to:
1. Ask natural language questions about an analyzed news article.
2. Receive precise, extractive answers grounded in the provided text.
3. Interact with the system using a transformer-based Question Answering pipeline.

```python
from transformers import pipeline

# Initialize the Extractive QA Pipeline
# Using a model trained on SQuAD (Stanford Question Answering Dataset)
qa_model = pipeline("question-answering", model="deepset/minilm-uncased-squad2")

print("✅ Conversational QA Engine Loaded Successfully.")
```

```python
# Define the Article Context (Knowledge Base)
context = """
Houston City College has announced a major expansion of its Artificial Intelligence and Robotics program. 
The college plans to invest $15 million in a new research facility by the summer of 2026. 
This expansion is designed to accommodate a 40% increase in student enrollment and will provide 
advanced labs for students to work on Natural Language Processing and Autonomous Systems.
"""

def ask_newsbot(question, article_context):
    """Processes a user query against the article context."""
    result = qa_model(question=question, context=article_context)
    return {
        "Question": question,
        "Answer": result['answer'],
        "Confidence Score": round(result['score'] * 100, 2)
    }

# Test queries to evaluate natural language interaction
queries = [
    "How much is the college investing?",
    "What is the deadline for the new facility?",
    "What specific labs will be provided?"
]

import pandas as pd
results_df = pd.DataFrame([ask_newsbot(q, context) for q in queries])
display(results_df)
```

## Findings
By utilizing the **MiniLM** transformer model, NewsBot 2.0 successfully moves from basic keyword searching to **Natural Language Interaction**. 
The system can pinpoint specific facts within the text and return them to the user with a confidence score, satisfying the requirement for **Natural Language Query Handling** and context-aware conversation.
