## Objective
This notebook demonstrates **Module C: Multilingual Intelligence**. 
We implement language detection, automated translation, and cross-lingual comparison to understand regional perspectives on global events.

```python
!pip install langdetect deep-translator --quiet
from langdetect import detect
from deep_translator import GoogleTranslator

def compare_coverage(en_text, es_text):
    """Compares English vs Spanish coverage (Cross-Language Analysis requirement)."""
    # Translate Spanish to English to normalize comparison
    translated_es = GoogleTranslator(source='es', target='en').translate(es_text)
    
    print("--- Cross-Language Comparison Output ---")
    print(f"Original EN: {en_text[:60]}...")
    print(f"Translated ES: {translated_es[:60]}...")
    
    # Calculate semantic overlap (simple keyword intersection)
    en_set = set(en_text.lower().split())
    es_set = set(translated_es.lower().split())
    overlap = en_set & es_set
    
    print(f"\nShared Strategic Keywords: {list(overlap)[:5]}")

# Example Data: Fed Rate Hikes
en_news = "The Federal Reserve is expected to maintain high interest rates through 2026."
es_news = "La Reserva Federal mantendrá las tasas de interés elevadas hasta el 2026."

compare_coverage(en_news, es_news)
```

## Findings
By implementing a 'Translate-and-Compare' workflow, NewsBot 2.0 successfully bridges the gap between regional news sources. This allows the system to identify shared strategic keywords across multiple languages automatically[cite: 147].
