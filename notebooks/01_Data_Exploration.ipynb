```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Create a sample dataset for demonstration
data = {
    'text': [
        "OpenAI released a new GPT model today focusing on reasoning.",
        "The Supreme Court is reviewing the new legal framework for digital liability.",
        "Markets rallied as the Federal Reserve kept interest rates steady.",
        "New breakthroughs in cancer research were published in the Science journal.",
        "El mercado de valores cerró al alza hoy tras el anuncio oficial."
    ],
    'category': ['Tech', 'Legal', 'Finance', 'Science', 'Finance']
}
df = pd.DataFrame(data)
print("✅ Dataset Loaded.")
```

```python
# Feature Engineering: Calculate sentence and word lengths
df['word_count'] = df['text'].apply(lambda x: len(x.split()))
df['char_count'] = df['text'].apply(len)

# Visualize Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['word_count'], bins=5, kde=True, color='blue')
plt.title('Distribution of Article Word Counts')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.show()

print("✅ Statistical Analysis Complete.")
```
