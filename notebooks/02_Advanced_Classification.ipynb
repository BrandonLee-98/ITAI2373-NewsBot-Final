## Objective
This notebook demonstrates the implementation of **Module A's Enhanced Classification Engine**. 
Key goals include:
1. Moving beyond rule-based heuristics to Zero-Shot Machine Learning.
2. Implementing multi-level categorization.
3. Generating and visualizing confidence scores for model transparency.

```python
import pandas as pd
from transformers import pipeline

# Initialize the Zero-Shot Classifier
# This model understands semantic context without specific training on labels
classifier = pipeline("zero-shot-classification", model="valhalla/distilbart-mnli-12-3")

candidate_labels = [
    "Legal & Politics", 
    "Technology & AI", 
    "Finance & Markets", 
    "Health & Science", 
    "Sports & Entertainment"
]

print("✅ Zero-Shot Model Loaded Successfully.")
```

```python
def test_classification(text):
    """Runs the text through the classifier and returns top result with score."""
    result = classifier(text, candidate_labels=candidate_labels)
    return {
        "Text": text[:50] + "...",
        "Top Category": result['labels'][0],
        "Confidence": round(result['scores'][0] * 100, 2)
    }

# Sample articles for validation
samples = [
    "The central bank raised interest rates by 25 basis points to combat rising inflation.",
    "A new study shows that CRISPR technology could potentially cure certain genetic disorders.",
    "The legislative committee is drafting a new bill regarding data privacy and user consent."
]

results_df = pd.DataFrame([test_classification(s) for s in samples])
display(results_df)
```

## Analysis of Results
By providing a confidence score, the NewsBot system can assess when it is 'unsure' about a category. 
In a production environment, scores below a certain threshold (e.g., 40%) could be flagged for human review, ensuring **System Integration** and data quality standards.
