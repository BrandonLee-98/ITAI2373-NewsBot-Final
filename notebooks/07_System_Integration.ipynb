## Objective
This final notebook demonstrates the **end-to-end integration** of all four modules. 
We will pass a raw, non-English news article through the following automated sequence:
1. **Language Detection & Translation** (Module C)
2. **Zero-Shot Classification & Confidence Scoring** (Module A)
3. **Abstractive Summarization** (Module B)
4. **Sentiment & Entity Extraction** (Module A)
5. **Contextual Question Answering** (Module D)

```python
# Final Setup: Import all logic from previously tested modules
from transformers import pipeline
from langdetect import detect
from deep_translator import GoogleTranslator
import pandas as pd

print("🚀 Initializing Unified Intelligence Pipeline...")

# Pre-loading models to simulate production startup
classifier = pipeline("zero-shot-classification", model="valhalla/distilbart-mnli-12-3")
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")
qa_bot = pipeline("question-answering", model="deepset/minilm-uncased-squad2")

labels = ["Finance", "Tech", "Politics", "Health", "Legal"]
print("✅ All NLP Engines Online.")
```

```python
def run_newsbot_pipeline(raw_text):
    """The Master Controller for the NewsBot Intelligence System."""
    report = {}
    
    # Step 1: Multilingual Pivot
    lang = detect(raw_text)
    report['Original Language'] = lang.upper()
    working_text = GoogleTranslator(source='auto', target='en').translate(raw_text) if lang != 'en' else raw_text
    
    # Step 2: Content Analysis
    cls_res = classifier(working_text, candidate_labels=labels)
    report['Category'] = cls_res['labels'][0]
    report['Confidence'] = f"{round(cls_res['scores'][0] * 100, 2)}%"
    
    # Step 3: Intelligence Generation
    summ_res = summarizer(working_text, max_length=45, min_length=20, do_sample=False)
    report['Executive Summary'] = summ_res[0]['summary_text']
    
    return report, working_text

# Test Input: A Spanish news snippet about AI regulation
input_article = """
La Unión Europea ha aprobado hoy la primera ley integral del mundo para regular la inteligencia artificial. 
Las nuevas normas buscan garantizar que los sistemas de IA utilizados en la UE sean seguros, transparentes 
y respetuosos con los derechos fundamentales. Las empresas que no cumplan podrían enfrentar multas millonarias.
"""

final_report, english_context = run_newsbot_pipeline(input_article)
print("--- FINAL INTELLIGENCE REPORT ---")
for key, val in final_report.items():
    print(f"{key}: {val}")
```

```python
# Step 4: Final Verification via Conversational Query
question = "What are the penalties for companies that do not comply?"
answer = qa_bot(question=question, context=english_context)

print(f"\n--- CONVERSATIONAL VERIFICATION ---")
print(f"User Query: {question}")
print(f"NewsBot Answer: {answer['answer']}")
print(f"Verification Confidence: {round(answer['score'] * 100, 2)}%")
```
