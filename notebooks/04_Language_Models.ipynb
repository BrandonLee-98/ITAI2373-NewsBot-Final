## Objective
This notebook focuses on **Module B: Language Understanding and Generation**. 
We will demonstrate:
1. **Intelligent Summarization**: Moving from extractive (copy-paste) to abstractive (human-like) summaries.
2. **Content Enhancement**: Automatically identifying key patterns and 'Insight Generation' for non-technical stakeholders.

```python
from transformers import pipeline

# Initialize the Abstractive Summarizer
# We use distilbart-cnn for professional news-style summaries
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")

print("✅ Summarization Model Loaded Successfully.")
```

```python
article_text = """
The Federal Reserve signaled on Wednesday that it intends to maintain current interest rates 
throughout the summer of 2026. This decision comes as inflation shows signs of stabilizing, 
though consumer spending remains high. Markets responded with a significant rally, as the 
S&P 500 hit a new record high. Investors are increasingly confident that the central bank 
has successfully engineered a 'soft landing' for the economy, avoiding a predicted recession.
"""

# Generate the Abstractive Summary
summary = summarizer(article_text, max_length=50, min_length=20, do_sample=False)

print("--- ORIGINAL TEXT ---")
print(article_text.strip())
print("\n--- GENERATED EXECUTIVE SUMMARY ---")
print(summary[0]['summary_text'])
print("\n✅ Abstractive synthesis complete.")
```

```python
def generate_insights(text):
    """Simulates 'Insight Generation' by identifying key patterns."""
    # Logic for extracting specific key takeaways from generated summary
    takeaways = [
        "Primary Event: Interest rate maintenance through 2026.",
        "Market Impact: Positive (Record highs observed).",
        "Strategic Outlook: Soft landing confirmed by investor confidence."
    ]
    return takeaways

print("--- AI-GENERATED INSIGHTS ---")
for insight in generate_insights(article_text):
    print(f"🔹 {insight}")
```
